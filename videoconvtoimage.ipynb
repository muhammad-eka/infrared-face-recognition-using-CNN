```python

import cv2
vidcap = cv2.VideoCapture('Video Generator/Fuad/RGB-Fuad berjalan.mp4') # visible images
vidcap2 = cv2.VideoCapture('Video Generator/Fuad/Thermal-Fuad berjalan.mp4') #thermal images
success,image = vidcap.read()
success2, image2 = vidcap2.read()
count = 14999
count2 = 14999
while success:
    cv2.imwrite("Image Extract From Video/Fuad/RGB/Berjalan/fuad.%d.jpg" % count, image)     # RGB save frame as JPEG file 
    cv2.imwrite("Image Extract From Video/Fuad/Thermal/Berjalan/fuad.%d.jpg" % count2, image2)     # Thermal save frame as JPEG file
    succes2,image2 = vidcap2.read()
    success,image = vidcap.read()
    print('Read a new thermal frame:', success2)
    print('Read a new rgb frame:', success)
    count += 1
    count2 +=1
```
